```python
import pandas as pd
```

```python
# Series (column of a dataframe, akin to a py dictionary)

values = [10,20,30,40,50]
s = pd.Series(values, index= ['a','b','c','d','e'])

s.loc['a'] # can look up an object by index

# DataFrame (concatenation of series) - will build a table with an automatic index

df = pd.DataFrame({
    'name' : ['Gam', 'Atarah', 'Shem'],
    'age': [37, 37, 7],
    'job':['Student', 'Teacher','Child']
})

df = df.set_index('name') #changes the index so that we can search with .loc[]

```

```python
df.loc['Gam']
```

```python
# by providing an index, we are able to do arithmatic operations of dataframes
df1 = pd.DataFrame({
    'a':[1,2,3]
},index=[0,1,2])

df2 = pd.DataFrame({
    'a':[10,20,30]
},index=[1,2,0])

df1 + df2
```

```python
# Can export the datafreame to other formats
# first reset the index so that the index is numerical instead of by name

df = df.reset_index()
```

```python
# by hitting the tab button after typing, the menu populates
# export to csv
df.to_csv('mydata.csv',index=None)
```

```python
pd.read_csv('mydata.csv')
```

```python
# when we read in the dataframe initially, we have an unnamed col (if the DF is not index=None). I can set the index_col to 0 
pd.read_csv('mydata.csv',index_col=0)
```
